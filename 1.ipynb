## Step 5: Compute and rank canton-level population density

```python
# First, we need to calculate total residents and total surface area per canton
canton_population_density = communes_data.groupby('Canton').agg(
    total_residents=('Residents', 'sum'),
    total_area=('Surface area in km²', 'sum')
)

# Then, compute population density at the canton level
canton_population_density['density'] = canton_population_density['total_residents'] / canton_population_density['total_area']

# The next step is to rank cantons by population density in descending order for clarity
canton_population_density = canton_population_density.sort_values(by='density', ascending=False)

# Then I want to reset index and add 1 to start indexing at 1
canton_population_density.reset_index(inplace=True)
canton_population_density.index = canton_population_density.index + 1

# Show Canton and Density only
canton_population_density[['Canton', 'density']]
```

# Probabilities

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('p1_communes_clean.csv')

# Inspect the dataset showing initial rows
print("Dataset Overview:")
print(df.head())

# First, filter communes with elevation > 2000
high_elevation_communes = df[df['Elevation'] > 2000]
total_high_elevation = high_elevation_communes.shape[0]
print(f"\nTotal communes with elevation over 2000: {total_high_elevation}")

# Second, count how many of these communes are from the canton of Valais (VS)
valais_communes_count = high_elevation_communes[high_elevation_communes['Canton'] == 'VS'].shape[0]
print(f"Number of high-elevation communes in Valais: {valais_communes_count}")

# Third, compute the probability
probability_valais = valais_communes_count / total_high_elevation if total_high_elevation > 0 else 0
print(f"\nProbability that a high-elevation commune is from Valais: {probability_valais:.2f}")
```

- Valais is home to several of Switzerland's highest peaks and alpine regions, making this result expected.

# Matrices

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('p1_communes_clean.csv')

# Use pd.crosstab to create the matrix with 0/1 encoding
canton_matrix = pd.crosstab(index=df['Commune'], columns=df['Canton'])

# Display the resulting matrix
print("Matrix of Communes and Cantons:")
print(canton_matrix.head())

# Save the matrix for future use
canton_matrix.to_csv('commune_canton_matrix.csv', index_label='Commune')
```

### My Observations:

A one-hot encoded matrix was created to map Swiss communes to their respective cantons. Rows represent communes, and columns represent cantons, with `1` indicating the canton they belong to.
This sparse matrix is easy to store and process because it only includes the necessary data and avoids repeating information.

- **Matrix Structure**: 
  - 26 columns (cantons) and rows for each commune.
  - Sparse matrix with a single `1` per row, as each commune belongs to exactly one canton.
- **Implementation**:
  - Used `pd.get_dummies()` for efficient one-hot encoding.
  - Indexed by commune names for clarity.
- **Applications**:
  - Enables canton-level grouping for statistical, geospatial, or machine learning analyses.

#### Example
The preview shows communes like "Aeugst am Albis" belonging to Zurich (`Canton_ZH`).
